## Event shuffling ##
We here present randomization techniques which shuffle events. All functions take a `snapshot_sequence()` object as input, and return a `snapshot_sequence()` object.

### Global permutation of events $\text{P}[E]$: `P__E(seq_data)` ###
This function randomly permutates events of a `snapshot_sequence()` object `seq_data`.

```python
seq_data = rn.read_snapshot_sequence("Bureau/Random/Library/timeset_InVS13.dat")
new_seq = rn.P__E(seq_data)
```

```python
TL0 = rn.activity_timeline(seq_data)
TL1 = rn.activity_timeline(new_seq)

tmin = 1000
tmax = 2000

fig = plt.figure(figsize=(8,2))
x = range(len(TL0))[tmin:tmax]
plt.plot(x,TL0[tmin:tmax],'k-')
plt.plot(x,TL1[tmin:tmax],'r-')
plt.show()
```
